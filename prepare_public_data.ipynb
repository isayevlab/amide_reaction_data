## Prepare reaction descriptors for fp and mordred

```python
folder = "/Users/liu5/Documents/data/datav6/datav6_public_internal"
rxns = rxn(folder)
ids = rxns.all_idx2()
```

```python
for i, id in enumerate(ids):
    fp = rxns.get_fp_fast(id)
    mordred = rxns.get_mordred_fast(id)
    aev = rxns.get_aev_fast(id)
    qm = rxns.get_qm_fast(id)
    steric = rxns.get_steric_desps(id)
    print(id)
    print(fp)
    print(mordred)
    print(aev)
    print(qm)
    print(steric)
    break
```

```python
print(len(ids))
fps, mordreds = [], []
for id in tqdm(ids):
    fps.append(rxns.get_fp(id))
    mordreds.append(rxns.get_mordred(id))
```

```python
fp_df = pd.DataFrame(fps, index=ids)
fp_df.head()
```

```python
fp_df.to_csv(os.path.join(folder, "reaction_fp_desps.csv"))
```

```python
mordred_df = pd.DataFrame(mordreds, index=ids)
mordred_df.head()
```
